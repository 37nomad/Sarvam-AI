```python
import os
import json
# Folder containing images
image_folder = "/kaggle/input/images"  # Change this to your folder path

# Get list of image files
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]
```

```python
len(image_files)
```

```python
image_files.sort()
```

```python
image_files
```

```python
# Store results
from tqdm import tqdm
ground_truth_data = []

for image_file in tqdm(image_files):
    image_path = os.path.join(image_folder, image_file)
    image = Image.open(image_path)
    
    # Extract text using Tesseract
    text = pytesseract.image_to_string(image, lang="eng")
    
    # Store image name and extracted text
    ground_truth_data.append({"image": image_file, "text": text})

# Save the entire dataset as a JSON file
output_json = "ground_truth.json"
with open(output_json, "w", encoding="utf-8") as file:
    json.dump(ground_truth_data, file, indent=4, ensure_ascii=False)

print(f"Ground truth saved to {output_json}")
```

```python
with open("ground_truth.json", "r", encoding="utf-8") as file:
    data = json.load(file)
```
